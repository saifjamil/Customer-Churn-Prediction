Impute order amount hike from last year

```python
# Step 1: Calculate median OrderAmountHike based on OrderCount & CashbackAmount
hike_median_by_group = df.groupby(["OrderCount"])["OrderAmountHikeFromlastYear"].median()

# Step 2: Define function to fill missing values
def impute_hike(row):
    if pd.isna(row["OrderAmountHikeFromlastYear"]):
        if row["Tenure"] < 2:  # New users likely have low hike
            return 0.0
        return hike_median_by_group.get(row["OrderCount"], df["OrderAmountHikeFromlastYear"].median())
    return row["OrderAmountHikeFromlastYear"]

# Apply imputation function
df["OrderAmountHikeFromlastYear"] = df.apply(impute_hike, axis=1)

# Step 3: Ensure no NaN values remain
df["OrderAmountHikeFromlastYear"].fillna(df["OrderAmountHikeFromlastYear"].median(), inplace=True)

print(df["OrderAmountHikeFromlastYear"].isnull().sum())  # Should print 0 if all missing values are handled
```

```python

### 1. Impute CouponUsed (based on OrderCount)
coupon_median_by_orders = df.groupby("OrderCount")["CouponUsed"].median()

def impute_coupon(row):
    if pd.isna(row["CouponUsed"]):
        return 0 if row["OrderCount"] == 0 else coupon_median_by_orders.get(row["OrderCount"], df["CouponUsed"].median())
    return row["CouponUsed"]

df["CouponUsed"] = df.apply(impute_coupon, axis=1)

### 2. Impute OrderCount (based on DaySinceLastOrder & other factors)
order_median_by_days = df.groupby("DaySinceLastOrder")["OrderCount"].median()
order_median_by_tenure = df.groupby("Tenure")["OrderCount"].median()

def impute_order(row):
    if pd.isna(row["OrderCount"]):
        # If the user hasn't ordered in a long time, assume low OrderCount
        if row["DaySinceLastOrder"] > 100:  
            return 1
        # Use DaySinceLastOrder as a reference for typical OrderCount
        if not pd.isna(row["DaySinceLastOrder"]):
            return order_median_by_days.get(row["DaySinceLastOrder"], df["OrderCount"].median())
        # As a fallback, use Tenure as a predictor (longer tenure → more orders)
        return order_median_by_tenure.get(row["Tenure"], df["OrderCount"].median())
    return row["OrderCount"]

df["OrderCount"] = df.apply(impute_order, axis=1)

### 3. Impute DaySinceLastOrder (based on OrderCount & Tenure)
day_median_by_orders = df.groupby("OrderCount")["DaySinceLastOrder"].median()
day_median_by_tenure = df.groupby("Tenure")["DaySinceLastOrder"].median()

def impute_days(row):
    if pd.isna(row["DaySinceLastOrder"]):
        # If never ordered, set a high value
        if row["OrderCount"] == 0:
            return 365  
        # Use OrderCount as a reference for typical inactivity periods
        if not pd.isna(row["OrderCount"]):
            return day_median_by_orders.get(row["OrderCount"], df["DaySinceLastOrder"].median())
        # Use Tenure to estimate inactivity if other data is missing
        return day_median_by_tenure.get(row["Tenure"], df["DaySinceLastOrder"].median())
    return row["DaySinceLastOrder"]

df["DaySinceLastOrder"] = df.apply(impute_days, axis=1)

# ✅ Final check for missing values
print("Missing values after imputation:")
print(df.isnull().sum())
```

```python
# Import necessary libraries
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Define the Transformer-based classifier
class TransformerClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_heads=4, num_layers=2, output_dim=1):
        super(TransformerClassifier, self).__init__()
        self.embedding = nn.Linear(input_dim, hidden_dim)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=num_heads),
            num_layers=num_layers
        )
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        # x shape: (batch_size, sequence_length, feature_dim)
        x = self.embedding(x)  # (batch_size, sequence_length, hidden_dim)
        x = x.permute(1, 0, 2)  # Transformer expects (sequence_length, batch_size, hidden_dim)
        x = self.transformer(x)  # (sequence_length, batch_size, hidden_dim)
        x = x.mean(dim=0)  # Pooling over the sequence dimension
        x = self.fc(x)  # (batch_size, output_dim)
        return self.sigmoid(x)

# Preprocess the data
X = df.drop('Churn', axis=1)  # Replace 'Churn' with your target column
y = df['Churn']

# One-hot encode categorical features if necessary
X = pd.get_dummies(X)

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)

# Define model, loss function, and optimizer
input_dim = X_train_tensor.shape[1]
model = TransformerClassifier(input_dim=input_dim)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# Add a sequence dimension to the input tensors
X_train_tensor_seq = X_train_tensor.unsqueeze(1)  # (batch_size, sequence_length=1, feature_dim)
X_test_tensor_seq = X_test_tensor.unsqueeze(1)

# Training loop
epochs = 100
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()

    # Forward pass
    output = model(X_train_tensor_seq)
    loss = criterion(output, y_train_tensor)

    # Backward pass and optimization
    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")

# Evaluate the model
model.eval()
with torch.no_grad():
    predictions = model(X_test_tensor_seq)
    predictions = (predictions > 0.5).float()
    accuracy = (predictions.eq(y_test_tensor).sum() / y_test_tensor.shape[0]).item()
    print(f"Test Accuracy: {accuracy:.4f}")
```

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# Define the Transformer-based classifier
class TransformerClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_heads=4, num_layers=2, output_dim=1):
        super(TransformerClassifier, self).__init__()
        self.embedding = nn.Linear(input_dim, hidden_dim)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=num_heads),
            num_layers=num_layers
        )
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.embedding(x)  # (batch_size, sequence_length, hidden_dim)
        x = x.permute(1, 0, 2)  # Transformer expects (sequence_length, batch_size, hidden_dim)
        x = self.transformer(x)  # (sequence_length, batch_size, hidden_dim)
        x = x.mean(dim=0)  # Pooling over the sequence dimension
        x = self.fc(x)  # (batch_size, output_dim)
        return self.sigmoid(x)

# Hybrid model combining XGBoost and Transformer
class XGBoostTransformer:
    def __init__(self, input_dim, hidden_dim=64, num_heads=4, num_layers=2):
        self.transformer_model = TransformerClassifier(input_dim, hidden_dim, num_heads, num_layers)
        self.xgboost_model = XGBClassifier(eval_metric='logloss')

    def fit(self, X_train, y_train):
        # Standardize the features
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)

        # Train the Transformer model
        X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32).unsqueeze(1)  # Add sequence dimension
        y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)

        criterion = nn.BCELoss()
        optimizer = optim.Adam(self.transformer_model.parameters(), lr=1e-3)

        # Training loop for Transformer
        epochs = 100
        for epoch in range(epochs):
            self.transformer_model.train()
            optimizer.zero_grad()
            output = self.transformer_model(X_train_tensor)
            loss = criterion(output, y_train_tensor)
            loss.backward()
            optimizer.step()

        # Get the Transformer predictions
        with torch.no_grad():
            transformer_predictions = self.transformer_model(X_train_tensor).numpy()

        # Train the XGBoost model on the Transformer predictions
        self.xgboost_model.fit(transformer_predictions, y_train)

    def predict(self, X_test):
        # Standardize the features
        scaler = StandardScaler()
        X_test_scaled = scaler.fit_transform(X_test)

        # Get Transformer predictions
        X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32).unsqueeze(1)  # Add sequence dimension
        with torch.no_grad():
            transformer_predictions = self.transformer_model(X_test_tensor).numpy()

        # Get final predictions from XGBoost
        return self.xgboost_model.predict(transformer_predictions)

    def evaluate(self, X_test, y_test):
        y_pred = self.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred)
        return accuracy, report
    
#evaluate
hybrid_model = XGBoostTransformer(input_dim=X_train.shape[1])
hybrid_model.fit(X_train, y_train)
accuracy, report = hybrid_model.evaluate(X_test, y_test)
print(f"Hybrid Model Accuracy: {accuracy:.4f}")
print("Classification Report:\n", report)
```
